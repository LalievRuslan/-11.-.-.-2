### **Задание 1:**

**Уравнение 1:** **$$y' - y = 2x - 3$$**

Это линейное дифференциальное уравнение первого порядка. Чтобы решить его, сначала найдем интегрирующий множитель. Интегрирующий множитель для уравнения вида $y' - p(x)y = q(x)$ определяется как $e^{-\int p(x)dx}$.

Для нашего уравнения $p(x) = -1$, поэтому интегрирующий множитель равен:
$$
\mu(x) = e^{-\int -1 dx} = e^{x}
$$

Умножим обе части уравнения на $\mu(x)$:
$$
e^{x}y' - e^{x}y = e^{x}(2x - 3)
$$

Теперь левая часть уравнения является производной от произведения $\mu(x)y$:
$$
\frac{d}{dx}(e^{x}y) = e^{x}(2x - 3)
$$

Интегрируем обе части уравнения по $x$:
$$
\int \frac{d}{dx}(e^{x}y) dx = \int e^{x}(2x - 3) dx
$$

Используем интеграл от произведения для правой части:
$$
e^{x}y = e^{x} \left( 2x - 2 \right) + Ce^{x}
$$

Теперь разделим обе части на $e^{x}$, чтобы найти общее решение уравнения:
$$
y = 2x - 2 + Ce^{-x}
$$
где $C$ — константа интегрирования.

**Уравнение 2:** **$$x^2 y' + xy + 1 = 0$$**

Это неоднородное линейное дифференциальное уравнение первого порядка. Для его решения применим метод замены переменной. Сначала перепишем уравнение в следующем виде:

$$
y' + \frac{1}{x}y = -\frac{1}{x^2}
$$

Теперь найдем интегрирующий множитель $\mu(x)$, который в данном случае равен $e^{\int \frac{1}{x}dx} = e^{\ln|x|} = x$.

Умножим обе части уравнения на интегрирующий множитель:

$$
x y' + y = -\frac{1}{x}
$$

Теперь левая часть уравнения является производной от произведения $xy$:

$$
(xy)' = -\frac{1}{x}
$$

Интегрируем обе части уравнения по $x$:

$$
\int (xy)' dx = \int -\frac{1}{x} dx
$$

$$
xy = -\ln|x| + C
$$

где $C$ — константа интегрирования. Теперь разделим обе части уравнения на $x$ (при условии, что $x \neq 0$):

$$
y = -\frac{\ln|x|}{x} + \frac{C}{x}
$$

Это общее решение дифференциального уравнения.

Обратите внимание, что при решении использовал свойство логарифма $\int \frac{1}{x} dx = \ln|x| + C$ и правило дифференцирования произведения функций $(fg)' = f'g + fg'$.

### **Задание 2:**

**Уравнение 1:**


Рассмотрим ряд с общим членом $a_n = \frac{n+2}{n^2 + n + 1}$ и исследуем его на сходимость.

**Предельный признак сравнения**

Предельный признак сравнения гласит, что если предел отношения двух последовательностей равен конечному числу, отличному от нуля, то оба ряда ведут себя одинаково (оба сходятся или оба расходятся). Для нашего ряда возьмем в качестве сравнения гармонический ряд $\sum \frac{1}{n}$, который является расходящимся. Исследуем предел:

$$
\lim_{{n \to \infty}} \frac{a_n}{\frac{1}{n}} = \lim_{{n \to \infty}} \frac{\frac{n+2}{n^2 + n + 1}}{\frac{1}{n}} = \lim_{{n \to \infty}} \frac{n(n+2)}{n^2 + n + 1}
$$

Упростим выражение, разделив числитель и знаменатель на $n^2$:

$$
= \lim_{{n \to \infty}} \frac{1 + \frac{2}{n}}{1 + \frac{1}{n} + \frac{1}{n^2}}
$$

При $n \to \infty$ члены с $n$ в знаменателе стремятся к нулю:

$$
= \frac{1 + 0}{1 + 0 + 0} = 1
$$

Поскольку предел отношения общего члена нашего ряда к общему члену гармонического ряда равен 1, и гармонический ряд расходится, то по предельному признаку сравнения наш ряд также расходится.

**Вывод**

Ряд $\sum_{n=1}^{\infty} \frac{n+2}{n^2 + n + 1}$ расходится.

**Уравнение 2:**


Рассмотрим ряд с общим членом $ a_n = \frac{n^n}{n!} $. Исследуем его на сходимость.

**Признак отношения (признак Даламбера)**

Признак отношения (признак Даламбера) утверждает, что если предел отношения последовательных членов ряда $ \lim_{{n \to \infty}} \frac{a_{n+1}}{a_n} $ стремится к числу $ L $, и если $ L < 1 $, то ряд сходится; если $ L > 1 $, то ряд расходится; если $ L = 1 $, то признак не даёт ответа.

Вычислим отношение $ \frac{a_{n+1}}{a_n} $:

$$
\frac{a_{n+1}}{a_n} = \frac{\frac{(n+1)^{n+1}}{(n+1)!}}{\frac{n^n}{n!}} = \frac{(n+1)^{n+1} \cdot n!}{n^n \cdot (n+1)!}
$$

Упростим выражение, сократив факториалы:

$$
= \frac{(n+1)^{n+1}}{n^n \cdot (n+1)} = \left(\frac{n+1}{n}\right)^n \cdot \frac{n+1}{n+1} = \left(1 + \frac{1}{n}\right)^n
$$

Теперь вычислим предел при $ n \to \infty $:

$$
\lim_{{n \to \infty}} \left(1 + \frac{1}{n}\right)^n = e
$$

Значение данного предела равно числу $ e $ (основание натурального логарифма), которое больше 1.

**Вывод**

Поскольку $ \lim_{{n \to \infty}} \frac{a_{n+1}}{a_n} = e > 1 $, то по признаку Даламбера ряд $ \sum_{n=1}^{\infty} \frac{n^n}{n!} $ расходится.

**Уравнение 3:**

Рассмотрим ряд с общим членом $ a_n = n \left(1 - \frac{1}{n}\right)^{n^2} $. Исследуем его на сходимость.

**Радикальный признак Коши**

Радикальный признак Коши гласит, что если существует предел $ \lim_{{n \to \infty}} \sqrt[n]{|a_n|} = q $, и если $ q < 1 $, то ряд сходится; если $ q > 1 $, то ряд расходится; если $ q = 1 $, то признак не даёт ответа.

Вычислим $ \sqrt[n]{|a_n|} $:

$$
\sqrt[n]{|a_n|} = \sqrt[n]{n \left(1 - \frac{1}{n}\right)^{n^2}} = \left(1 - \frac{1}{n}\right)^n \cdot \sqrt[n]{n}
$$

Теперь вычислим предел при $ n \to \infty $:

$$
\lim_{{n \to \infty}} \left(1 - \frac{1}{n}\right)^n = \frac{1}{e}
$$

и

$$
\lim_{{n \to \infty}} \sqrt[n]{n} = 1
$$

Таким образом, предел радикала Коши равен:

$$
q = \frac{1}{e} \cdot 1 = \frac{1}{e}
$$

Так как $ \frac{1}{e} < 1 $, ряд сходится по радикальному признаку Коши.

**Вывод**

Ряд $ \sum_{n=1}^{\infty} n \left(1 - \frac{1}{n}\right)^{n^2} $ сходится.

**Уравнение 4:**

Рассмотрим ряд с общим членом $ a_n = \frac{(-1)^{n+1}}{2n - \ln n} $. Исследуем его на сходимость.

**Признак Лейбница для знакочередующихся рядов**

Признак Лейбница гласит, что знакочередующийся ряд $ \sum_{n=1}^{\infty} (-1)^{n} a_n $, где $ a_n > 0 $, сходится, если выполняются два условия:

1. Члены ряда монотонно убывают: $ a_{n+1} \leq a_n $.
2. Члены ряда стремятся к нулю: $ \lim_{{n \to \infty}} a_n = 0 $.

Сначала убедимся, что члены ряда положительны и знакочередующиеся:

$$
a_n = \frac{(-1)^{n+1}}{2n - \ln n}
$$

Для $ n \geq 1 $, знаменатель $ 2n - \ln n > 0 $, так как $ 2n > \ln n $. Следовательно, $ a_n $ положительно при четных $ n $ и отрицательно при нечетных $ n $, что соответствует знакочередованию.

Теперь проверим монотонность:

$$
a_{n+1} = \frac{1}{2(n+1) - \ln (n+1)}
$$

и

$$
a_n = \frac{1}{2n - \ln n}
$$

Нужно показать, что $ a_{n+1} < a_n $ для всех $ n $. Это эквивалентно утверждению, что $ 2n - \ln n < 2(n+1) - \ln (n+1) $, что верно, так как функция $ 2n - \ln n $ монотонно возрастает на $ n \geq 1 $.

Наконец, проверим, что члены ряда стремятся к нулю:

$$
\lim_{{n \to \infty}} a_n = \lim_{{n \to \infty}} \frac{(-1)^{n+1}}{2n - \ln n} = 0
$$

так как знаменатель растет быстрее, чем числитель.

**Вывод**

Так как оба условия признака Лейбница выполняются, ряд $ \sum_{n=1}^{\infty} \frac{(-1)^{n+1}}{2n - \ln n} $ сходится.

**Уравнение 5:**

Рассмотрим ряд с общим членом $ a_n = \frac{(x - 2)^{n+1}}{3^n (n + 2)} $. Исследуем его на сходимость.

**Признак Даламбера (отношения)**

Признак Даламбера гласит, что если предел отношения последовательных членов ряда $ \lim_{{n \to \infty}} \frac{a_{n+1}}{a_n} $ стремится к числу $ L $, и если $ L < 1 $, то ряд сходится; если $ L > 1 $, то ряд расходится; если $ L = 1 $, то признак не даёт ответа.

Вычислим отношение $ \frac{a_{n+1}}{a_n} $:

$$
\frac{a_{n+1}}{a_n} = \frac{\frac{(x - 2)^{n+2}}{3^{n+1} (n + 3)}}{\frac{(x - 2)^{n+1}}{3^n (n + 2)}} = \frac{(x - 2) (n + 2)}{3 (n + 3)}
$$

Теперь вычислим предел при $ n \to \infty $:

$$
\lim_{{n \to \infty}} \frac{a_{n+1}}{a_n} = \lim_{{n \to \infty}} \frac{(x - 2) (n + 2)}{3 (n + 3)} = \frac{x - 2}{3}
$$

Полученное значение $ \frac{x - 2}{3} $ — это радиус сходимости степенного ряда. Ряд сходится, если $ \left|\frac{x - 2}{3}\right| < 1 $, то есть:

$$
-1 < \frac{x - 2}{3} < 1
$$

Решая это неравенство, получаем:

$$
1 < x < 5
$$

**Вывод**

Ряд $ \sum_{n=1}^{\infty} \frac{(x - 2)^{n+1}}{3^n (n + 2)} $ сходится, если $ x $ принадлежит интервалу $ (1, 5) $.
